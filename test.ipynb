```python
def height(step, alter, octave):
    # A0 is 1
    height = 0
    if octave == 0:
        if step == 'A':
            height = 0 if alter == 0 else 1
        if step == 'B':
            if alter == 0:
                height = 3
            else:
                height = 2 if alter == -1 else 4
                
    d_note = {
        'C': 4,
        'D': 6,
        'E': 8,
        'F': 9,
        'G': 11,
        'A': 13,
        'B': 15
    }
    height = d_note[step] + (octave-1) * 12 + alter
    return height
```

```python
print height('C', 0, 1) # C1
print height('D', -1, 2) # bD2
print height('E', 1, 3) # #E3
print height('F', -1, 4) # bF4
print height('G', 1, 5) #  #G5
print height('A', -1, 6) # bA6
print height('B', 1, 7) # #B7
print height('C', 0, 8) # C8
print height('A', 0, 0) # A0
```
